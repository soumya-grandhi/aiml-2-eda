```python
import matplotlib.pyplot as plt
import seaborn as sns

data.hist(bins=20, figsize=(15,10), edgecolor='black')
plt.tight_layout()
plt.savefig("task2aiml_outputs/histograms.png")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=data[num_cols],orient='h')
plt.title("Boxplot for numerical features")
plt.savefig("task2aiml_outputs/boxplot.png")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True,fmt='.2f',cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("task2aiml_outputs/correlation_matrix.png")
plt.show()
```

```python
pairplot = sns.pairplot(data[num_cols])
pairplot.fig.savefig("task2aiml_outputs/pairplot.png")
plt.show()
```

```python
for col in num_cols:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3-q1
    outliers = data[(data[col]<q1-1.5*iqr) | (data[col]>q3+1.5*iqr)]
    if not outliers.empty:
        print(f"{col} has outliers: ")
        print(outliers[[col]])
              
```

```python
# High correlation? → Might cause multicollinearity.

# Skewed distribution? → Might need transformation (log, square root, etc.).

# Outliers in boxplot? → May need treatment depending on the context.

# Separation in pairplots by target? → Feature is informative for classification.

# Clustered patterns in scatter plots? → Indicates potential grouping.
```
